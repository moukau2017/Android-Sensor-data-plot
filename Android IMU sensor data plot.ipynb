### Sitting position

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_data(ax, prefix, title, ylabel):
    line_styles = ['-.', '--', '-']
    marker_styles = [None, 'h', None]
    line_widths = [2.5,1.5, 2.2]
    colors = ['#A52A2A', '#5F9EA0', '#008000']

    for i, axis in enumerate(['x', 'y', 'z']):
        column_name = f'{prefix}.{axis}'
        ax.plot(data[column_name], label=f'{axis.upper()}-axis', 
                linestyle=line_styles[i], marker=marker_styles[i], 
                linewidth=line_widths[i], color=colors[i])
    ax.set_title(title)
    ax.set_xlabel('Time')
    ax.set_ylabel(ylabel)
    ax.legend()

# Read the CSV file
data = pd.read_csv('sitting.csv')

# Create a new figure with 4 subplots
fig, axs = plt.subplots(4, 1, figsize=(10, 15))

# Plotting the data
plot_data(axs[0], 'lsm6dso_accelerometer', 'LSM6DSO Accelerometer', 'Acceleration (m/s^2)')
plot_data(axs[1], 'lsm6dso_gyroscope', 'LSM6DSO Gyroscope', 'Angular Velocity (/s)')
plot_data(axs[2], 'tmd4912_light_cct', 'TMD4912 Light', 'Illuminance (lx)')
plot_data(axs[3], 'ak09918_magnetometer', 'AK09918 Magnetometer', 'Orientation ()')

# Adjust the layout and show the plots
plt.tight_layout()
plt.show()
fig.savefig('sitting1.png')
```
